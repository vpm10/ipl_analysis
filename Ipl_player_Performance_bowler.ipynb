```python
season_played = data['Player'].value_counts().reset_index()
season_played.columns = ['player' , 'season_played']
season_played.sort_values('player',inplace=True)
season_played.reset_index(inplace=True)
season_played.drop(('index'),axis=1,inplace = True)
season_played.head()
```

### most seasons played 

```python
temp = season_played.sort_values('season_played' ,  ascending = False)[:100]
plt.figure(figsize = (26,8))
sns.barplot(data=temp,x='player',y='season_played')
plt.xticks(rotation='vertical')
plt.xlabel('Player Name')
plt.ylabel('Number of seasons played')
plt.show()
```

### most matches played 

```python
matches = data.groupby('Player')['Mat'].sum().reset_index()
matches.columns = ['player' , 'matches played']
matches.sort_values('player',inplace=True)
matches.reset_index(inplace=True)
matches.drop(('index'),axis=1,inplace = True)
matches.head()
```

```python
temp = matches.sort_values('matches played' ,  ascending = False)[:100]
plt.figure(figsize = (26,8))
sns.barplot(data=temp,x='player',y='matches played',palette='Paired')
plt.xticks(rotation='vertical')
plt.xlabel('Player Name')
plt.ylabel('Number of matches played')
plt.show()
```

```python
matches[matches['player']=='Ravindra Jadeja'] #most matches played 
```

```python
wkts = data.groupby('Player')['Wkts'].sum().reset_index()
wkts.columns = ['player' , 'wkts taken']
wkts.sort_values('player',inplace=True)
wkts.reset_index(inplace=True)
wkts.drop(('index'),axis=1,inplace = True)
temp = wkts.sort_values('wkts taken' ,  ascending = False)[:100]
plt.figure(figsize = (26,8))
sns.barplot(data=temp,x='player',y='wkts taken',palette='Paired')
plt.xticks(rotation='vertical')
plt.xlabel('Player Name')
plt.ylabel('Number of wickets taken')
plt.show()

```

### player with most over bowled
